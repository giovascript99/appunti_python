📝 Appunti Python: Variabili e Dati
🔹 Cos'è una Variabile?
Una variabile è un contenitore con un nome che punta a un valore.
In Python non serve dichiarare il tipo (es: "questo è un numero"), basta assegnare il valore con l'uguale =.

📦 Tipi di Dato Semplici
Integer (int) Numeri interi, senza decimali.

Esempio: eta = 25 o temperatura = -5

Float (float) Numeri decimali (si usa il punto, non la virgola).

Esempio: prezzo = 19.99

String (str) Testo racchiuso tra virgolette (" ") o apici (' ').

Esempio: nome = "Davide"

Boolean (bool) Valori logici. Esistono solo due stati: True o False.

Nota: In Python si scrivono sempre con la iniziale maiuscola.

Complex (complex) Numeri complessi usati in matematica avanzata. Sono composti da una parte reale e una immaginaria (indicata con la j).

Esempio: numero = 3 + 5j

🛠️ Esempi di Codice

```python
# Assegnazione
nome = "Studente Aulab"
eta = 25
altezza = 1.75
is_coding = True

# Stampiamo i tipi per verificare
print(type(nome))    # <class 'str'>
print(type(eta))     # <class 'int'>
print(type(is_coding)) # <class 'bool'>
```

⚠️ Regole per i Nomi delle Variabili
Per evitare errori, ricorda queste regole:

✅ Sì: nome_utente, voto2, _segreto

❌ No: 2nome (non può iniziare con un numero)

❌ No: nome-utente (il trattino è un meno!)

❌ No: class (non puoi usare parole riservate del linguaggio)

```python
# Escape character "\" -> andando ad aggiungere il \ alle "" diciamo a python che deve ignorare il significato sintattico di quello specifico carattere
message = "Egli disse: \"Che bella giornata!\""
print(message)

# \n crea una nuova riga
lista_spesa = "Pane\nLatte\nUova"
print(lista_spesa)

# \t crea un rientro (tabulazione)
paragrafo = "\tQuesto testo è rientrato."
print(paragrafo)

# Esempio combinato
messaggio_complesso = "Corso Python:\n\t- Variabili\n\t- Tipi di dato"
print(messaggio_complesso)

# Il "Trucco" delle triple virgolette
messaggio_lungo = """Questo è un testo
che va a capo da solo
senza dover usare caratteri di escape!"""
print(messaggio_lungo)
```

```python
# Metodo classico

name = "Davide"
age = 32
print("Ciao a tutti, mi chiamo", name, "ed ho", age, "anni")
```

```python
# String Interpolation -> processo tramite il quale dei segnaposto vengono sostituiti con i valori delle variabili corrispondenti

name = "Luca"
age = 39
print("Ciao a tutti, mi chiamo {0} ed ho {1} anni".format(name, age)) #possiamo anche indicare l'ordine in cui vengono utilizzate le variabili all'interno del metodo .format
```

```python
# Format Strings

name = "Giovanni"
age = 26
print(f"Ciao a tutti, mi chiamo {name} ed ho {age} anni")
print(f"L'anno prossimo avrò {age + 1} anni")

prezzo = 19.991234
print(f"Il prezzo è: {prezzo:.2f}€") # Output: 19.99€
```
