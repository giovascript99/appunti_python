TIPI DI DATO LISTA

```python
numbers = [10, 20, 30, 40]

# Attraverso la funzione len possiamo valutare la lunghezza di una lista
len(numbers)
```

```python
# Ogni elemento di una lista ha un indice associato che rappresenta la posizione all'interno della lista stessa
# Python è un linguaggio di programmazione Zero Based dove gli indici partono da 0
numbers = [10, 20, 30, 40]

print(numbers[1])
```

```python
# Le liste possono contenere tipi di dato diversi tra loro
list = [10, "Davide", True]

# Posso inserire all'interno delle mie liste anche altre liste (MATRICI)
teachers = [
    ["Davide", "Cariola"],
    ["Francesco", "Talamona"],
    ["Marco", "Insabato"]
]

print(teachers[1][1])
```

TIPI DI DATO TUPLA

```python
# Parentesi tonde
numbers = (10, 20, 30)

print(numbers[1])
```

```python
numbers = (10, 20, 30)

x, y, z = numbers

print(x)
print(y)
print(z)
```

```python
# Swap di due variabili
x = 5
y = 10

# temp -> variabile d'appoggio
temp = x
x = y
y = temp

print(x, y)
```

```python
# Swap di due variabili con la Tupla
x = 5
y = 10

x, y = (y, x)

print(x, y)
```

TIPI DI DATO DICTIONARY - Coppie Chiave -> Valore

```python
teacher = {
    'name' : 'Davide',
    'surname' : 'Cariola',
    'age' : 32
}

# square bracket notation
print(teacher["surname"])

# metodo get (metodo più moderno)
print(teacher.get('age'))
```

```python
teacher = {
    'name' : 'Davide',
    'surname' : 'Cariola',
    'age' : 32,
    'contract' : {
        'duration' : 'infinite',
        'remote' : True
    }
}

print(teacher.get('contract').get('duration'))
```
