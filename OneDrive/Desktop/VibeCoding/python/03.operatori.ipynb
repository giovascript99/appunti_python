```text
1. Operatori Aritmetici (Matematica)

+   Addizione             (10 + 5  -> 15)
-   Sottrazione           (10 - 5  -> 5)
*   Moltiplicazione       (10 * 5  -> 50)
/   Divisione             (10 / 4  -> 2.5)
//  Divisione Intera      (10 // 4 -> 2)
%   Modulo (Resto)        (10 % 3  -> 1)
**  Potenza               (2 ** 3  -> 8)


2. Operatori di Assegnazione

=   Assegnazione          (x = 5)               # Assegna 5 a x
+=  Somma e assegna       (x += 3 -> x = 8)     # x era 5, ora è 8
-=  Sottrai e assegna     (x -= 2 -> x = 6)     # x era 8, ora è 6
*=  Moltiplica e assegna  (x *= 2 -> x = 12)    # x era 6, ora è 12
/=  Dividi e assegna      (x /= 4 -> x = 3.0)   # x era 12, ora è 3.0


3. Operatori di Confronto (Risultato: True o False)

==  Uguale a              (5 == 5 -> True)      # Verifica se sono identici
!=  Diverso da            (5 != 3 -> True)      # Verifica se sono diversi
>   Maggiore              (5 > 10 -> False)     # 5 è più grande di 10?
<   Minore                (5 < 10 -> True)      # 5 è più piccolo di 10?
>=  Maggiore o uguale     (5 >= 5 -> True)      # 5 è maggiore o uguale a 5?
<=  Minore o uguale       (5 <= 2 -> False)     # 5 è minore o uguale a 2?


4. Operatori Logici

and  Entrambe vere        (5 > 3 and 2 > 1) -> True   # Entrambe devono essere True
or   Almeno una vera      (5 > 3 or 2 < 1)  -> True   # Basta che una sia True
not  Inverte il risultato (not(5 > 3))      -> False  # Trasforma True in False


5. Operatori di Appartenenza

in      Presente in       ("a" in "mela")     -> True   # Cerca "a" dentro "mela"
not in  NON presente in   ("z" not in "mela") -> True   # Verifica che "z" manchi


6. Operatori Bitwise (Lavoro sui Bit)

&   AND         (5 & 1 -> 1)      # 0101 & 0001 = 0001
|   OR          (5 | 1 -> 5)      # 0101 | 0001 = 0101
^   XOR         (5 ^ 1 -> 4)      # 0101 ^ 0001 = 0100 (bit diversi)
~   NOT         (~5    -> -6)     # Inverte tutti i bit
<<  Shift SX    (5 << 1 -> 10)    # Sposta i bit a sinistra (moltiplica x2)
>>  Shift DX    (5 >> 1 -> 2)     # Sposta i bit a destra (divide x2)
