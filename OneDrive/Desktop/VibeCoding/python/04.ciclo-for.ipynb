```python
# posso dire al mio codice di fermarsi al raggiungimento di una condizione con il break
min = -5
max = 10
counter = 0

for i in range(min, max + 1):
    if i % 2 == 0:
        print(i, 'pari')
        counter += 1
        
    if counter >= 5:
        break
```

```python
# break -> interrompe il ciclo immediatamente
for n in range(10):
    if n == 5:
        break
    print(n) # Stampa solo fino a 4
```

```python
# continue -> salta l'iterazione attuale e passa alla prossima
for n in range(5):
    if n == 2:
        continue
    print(n) # Stampa 0, 1, 3, 4 (salta il 2)
```
