```python
# Il while si basa su un'espressione booleana, se true il ciclo comincerà o continuerà. Dobbiamo indicare sia l'indice sia il suo metodo di incremento
i = 1

while i <= 10:
    print(i)
    i += 1
```

```python
n = 0
while n < 10:
    n += 1
    
    if n == 3:
        continue  # Salta il resto del codice e torna all'inizio del ciclo
        
    if n == 7:
        break     # Esce immediatamente dal ciclo
        
    print(n) # Stampa: 1, 2, 4, 5, 6
```
