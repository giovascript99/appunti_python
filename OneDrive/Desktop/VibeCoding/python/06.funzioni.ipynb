```python
# parametri
x = 5
y = 10

def sum(a, b):
    print(a + b)
    
sum(x, y)
```

```python
x = 5
y = 10

def divide(a, b):
    print(a / b)
    
divide(x, y)
divide(y, x)
```

```python
# Parametro di default -> scatterà in caso non venisse passato il numero giusto di parametri reali
x = 5
y = 10

def sum(a, b = 80):
    print(a + b)
    
sum(x, y)
sum(x)
```

```python
# Keyword arguments -> possiamo inviare degli argomenti ad una funzione tramite una sintassi chiave => valore
x = 5
y = 10

def divide(a, b):
    print(a / b)
    
divide(b = x, a = y)
```

```python
# Unpack Operator -> Si usa in quei rari casi in cui non sappiamo quanti parametri verranno passati all'interno della nostra funzione
x = 5
y = 10
z = 20

def sommatoria(*numbers):
    result = 0
    for number in numbers:
        result += number
    print(result)
    
sommatoria(x, y, z)
```

```python
# return statement
x = 5
y = 10
z = 20

def sommatoria(*numbers):
    result = 0
    for number in numbers:
        result += number
    return result
    
func_result = sommatoria(x, y, z)

print(func_result)
```
