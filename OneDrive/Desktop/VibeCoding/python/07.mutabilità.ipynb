L'immutabilità può essere usata per assicurarsi che un oggetto rimanga costante durante l'esecuzione di un algoritmo ed è particolarmente importante per data analysis e cybersecurity.

OGGETTI MUTABILI: liste, dizionari, set
OGGETTI IMMUTABILI: numeri, stringhe, tuple, booleani

```python
x = 500

print(x)
print(id(x)) # indirizzo di memoria
print(hex(id(x))) # indirizzo in valore esadecimale
```

```python
a = 500
b = 27

print(id(a), id(b))
```

```python
# Object Interning
num1 = 10
num2 = 10

print(id(num1), id(num2))
print(num1 is num2)
```

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print(hex(id(list1)))

list1[0] = 20
# Nonostante io dia un nuovo valore all'elemento 0 della lista, l'id della lista resta lo stesso quindi è mutabile
print(list1 is list2)
print(list1)
print(hex(id(list1)))
print(hex(id(list1[0])))
```

```python
string = 'Hello world!'

print(hex(id(string)))

string = 'Farewell world!'

print(hex(id(string)))

string[-1] = '?'
```
