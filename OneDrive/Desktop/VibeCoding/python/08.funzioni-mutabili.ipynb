🧬 Mutabilità nelle Funzioni Python

In Python, il modo in cui una funzione modifica un dato dipende dal fatto che il dato sia Mutabile (può cambiare) o Immutabile (non può cambiare).

1. Oggetti Immutabili (Numeri, Stringhe, Tuple)

Quando passi un dato immutabile a una funzione, Python ne crea una "copia" locale. Se lo modifichi dentro la funzione, l'originale fuori non cambia.

```python
def increment(n):
    n += 1
    
x = 5
increment(x)
print(x) # resta invariato
```

2. Oggetti Mutabili (Liste, Dizionari, Set)

Quando passi un oggetto mutabile, la funzione lavora sull'oggetto originale. Se lo modifichi all'interno, le modifiche saranno visibili anche all'esterno.

```python
def edit_name(dict):
    dict['first_name'] = "Roberto"
    
teacher = {'first_name': 'Davide', 'last_name': 'Cariola'}

edit_name(teacher)
print(teacher)
```

3. La tupla è un oggetto immutabile ma in alcuni casi è possibile modificare gli oggetti in essa contenuti, se l'oggetto contenuto è una Stringa, quindi immutabile, allora la tupla non può essere modificata e stiamo parlando di Immutabilità Non Transitiva, ma un campo mutabile come una Lista può essere modificato.

```python
# Esempio di Tupla immutabile -> da errore
tuple = ('Davide', 'Roberto')
print(hex(id(tuple)))

tuple = ('Paola', 'Marco')
print(hex(id(tuple)))

tuple[1] = 'Francesco'
print(hex(id(tuple)))
```

```python
# Esempio di Tupla mutabile
tuple = (1, 2, ['Hello', 'World'])
print(hex(id(tuple)))

tuple[2].append('!')
print(tuple)
print(hex(id(tuple)))
```
