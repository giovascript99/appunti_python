🏛️ Classi e Istanze in Python

Il "Progetto" e l' "Oggetto Reale": Fondamenti della Programmazione a Oggetti (OOP)

In Python, una Classe è come il progetto architettonico di una casa: definisce dove saranno le finestre e le porte, ma non è una casa vera. Un'Istanza è la casa fisica costruita seguendo quel progetto. Puoi costruire infinite case (istanze) partendo dallo stesso progetto (classe), ognuna con il proprio colore o indirizzo (attributi).

```python
class Person:
    counter = 0
    
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
        Person.counter += 1
        
    def say_hello(self):
        print(f"Mi chiamo {self.name} {self.surname} e ho {self.age} anni!")
        
    @classmethod # decoratore
    def counter_check(cls):
        print(f"Sinora ho creato {Person.counter} persone!")
        
Person.counter_check()
print("counter", Person.counter)

davide = Person('Davide', 'Cariola', 32)
roberto = Person('Roberto', 'Russo', 33)

Person.counter_check()

print("counter", Person.counter)

print(davide.name)
print(roberto.surname)

print(hex(id(davide.counter)))
print(hex(id(Person.counter)))

davide.say_hello()
```
