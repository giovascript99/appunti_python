🧬 Ereditarietà in Python

L'ereditarietà permette di creare nuove classi basate su classi esistenti, favorendo il riutilizzo del codice.

```python
class Person:
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
        
    def say_hello(self):
        print(f"Mi chiamo {self.name} {self.surname} e ho {self.age} anni!")
        
class Student(Person):
    def __init__(self, name, surname, age, subject):
        super().__init__(name, surname, age)
        self.subject = subject
        
    def say_hello(self):
        print(f"Ciao a tutti sono {self.name} {self.surname} e mi piace molto studiare {self.subject}")
        
        
mario = Student('Mario', 'Rossi', 20, 'Python')

mario.say_hello()

print(isinstance(mario, Student))
print(isinstance(mario, Person))
```

```python
class Teacher(Person):
    def __init__(self, name, surname, age, subject):
        super().__init__(name, surname, age)
        self.subject = subject
        
    def teacher_greet(self):
        print(f"Salve ragazzi sono {self.name} {self.surname} e oggi vedremo insieme {self.subject}")
            

davide = Teacher('Davide', 'Cariola', 32, 'Python')

davide.teacher_greet()
```

```python
print(issubclass(Student, Person))
print(issubclass(Teacher, Person))
```

 Ereditarietà Multipla

Python permette a una classe di ereditare da più classi contemporaneamente. La classe figlia avrà accesso ai metodi di tutti i suoi genitori.

```python
class Volante:
    def muovi(self):
        print("Sto volando in alto")

class Nuotante:
    def nuota(self):
        print("Sto nuotando in profondità")

# Drone eredita da entrambi
class DroneMarino(Volante, Nuotante):
    def missione(self):
        print("Inizio scansione area...")

robot = DroneMarino()
robot.muovi()  # Ereditato da Volante
robot.nuota()  # Ereditato da Nuotante
```
