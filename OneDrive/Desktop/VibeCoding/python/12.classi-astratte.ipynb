🏛️ Classi Astratte in Python (ABC)

Una Classe Astratta è una classe che non può essere istanziata (non puoi creare un oggetto direttamente da lei). Il suo unico scopo è servire come "modello" per altre classi.

```python
# abc -> Abstract Base Classes
from abc import ABCMeta, abstractmethod

class Person(metaclass=ABCMeta):
    @abstractmethod 
    def say_hello(self, args):
        pass
```

```python
class Teacher(Person):
    def __init__(self, name, surname, age, subject):
        self.name = name
        self.surname = surname
        self.age = age
        self.subject = subject
        
    def say_hello(self):
        print(f"Salve a tutti mi chiamo {self.name} {self.surname} ho {self.age} anni e oggi vedremo insieme {self.subject}")
        
davide = Teacher('Davide', 'Cariola', 32, 'Python')
davide.say_hello()
```
