🧩 Composizione di Oggetti in Python

La Composizione è un principio del design orientato agli oggetti in cui una classe complessa viene costruita utilizzando istanze di altre classi come componenti.

Mentre l'ereditarietà definisce una relazione di tipo "è un" (is-a), la composizione definisce una relazione di tipo "ha un" (has-a).

1. Ereditarietà vs Composizione

Ereditarietà (is-a): Un Insegnante è una Persona. Eredita tutte le caratteristiche, anche quelle che potrebbero non servire.

Composizione (has-a): Un Automobile ha un Motore. L'automobile non "è" un motore, ma lo contiene per funzionare.

```python
from abc import ABCMeta, abstractmethod

class MotherBoard(metaclass=ABCMeta):
    pass

class MsiB550(MotherBoard):
    def __init__(self):
        self.format = 'ATX'
        self.socket = 'AM4'
        self.ram = 'DDR4'
        
class MsiCreator(MotherBoard):
    def __init__(self):
        self.format = 'E-ATX'
        self.socket = 'TRX4'
        self.ram = 'DDR4'
        
class Computer:
    def __init__(self, motherBoard):
        self.motherBoard = motherBoard
        
    def details(self):
        print(f"Formato: {self.motherBoard.format}; Socket: {self.motherBoard.socket}; RAM: {self.motherBoard.ram}\n")
        
computer1 = Computer(MsiB550())
computer2 = Computer(MsiCreator())

computer1.details()
computer2.details()
```

🔍 Spiegazione Semplice: Cosa abbiamo fatto?

Immagina di costruire un PC reale. Non compri un blocco unico saldato, ma scegli i pezzi separatamente. Il codice sopra riflette esattamente questo processo:

L'Interfaccia (MotherBoard): Abbiamo creato un "marchio di garanzia". Dicendo che le schede MSI sono di tipo MotherBoard, assicuriamo che siano compatibili con il nostro computer.

I Componenti (MsiB550, MsiCreator): Queste sono le scatole dei pezzi hardware. Ogni scheda ha le sue caratteristiche (Socket AM4 o TRX4). Nota che queste classi non sanno nulla del computer, sanno solo cosa c'è scritto sulla loro scatola.

La Composizione (Computer): Questa è la parte magica. Nel metodo __init__, la classe accetta un pezzo esterno (motherBoard). Invece di "essere" una scheda madre (Ereditarietà), il computer ne contiene una.

Quando scriviamo self.motherBoard = motherBoard, stiamo letteralmente inserendo la scheda nel case del PC.

Il Polimorfismo in azione: Nel metodo details(), il computer dice: "Ehi scheda madre (qualunque tu sia), dimmi che socket hai". Il computer non sa se dentro c'è una B550 o una Creator, gli basta sapere che l'oggetto che gli abbiamo passato ha un attributo .socket.

🌟 Perché è meglio così?

Se domani esce una scheda madre della Asus, non devi cambiare il codice della classe Computer. Crei semplicemente la classe Asus e la passi al computer. Il Computer è flessibile: accetta qualsiasi pezzo purché "incastri" bene.
