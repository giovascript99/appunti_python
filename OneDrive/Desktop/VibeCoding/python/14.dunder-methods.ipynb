✨ Metodi Dunder (Magic Methods)

I metodi Dunder (abbreviazione di Double Underscore) sono metodi speciali che iniziano e finiscono con due trattini bassi, come __init__. Permettono di definire come i tuoi oggetti devono comportarsi con gli operatori standard di Python.

```python
class Person:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        
    def __str__(self):
        return f"{self.name} {self.surname}"
    
    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.surname == other.surname
        else:
            raise Exception(f"{other} non è un'istanza di Person")
        
    def __setitem__(self, key, value):
        if key not in ['name', 'surname']:
            raise Exception(f"{key} non è valida per la classe Person")
        elif key == 'name':
            self.name = value
        elif key == 'surname':
            self.surname = value
            
    def __len__(self):
        return len(self.name) + len(self.surname)
    
    def __add__(self, other):
        if isinstance(other, ComplexNumber):
            return ComplexNumber(self.real + other.real, self.imaginary + other.imaginary)
        else:
            raise Exception(f"{other} non è un tipo di dato compatibile con questa operazione")
        
        
davide = Person('Davide', 'Cariola')
davide2 = Person('Davide', 'Cariola')

print(davide)
print(davide == davide2)
print(len(davide))
```

1. __init__ (L'Inizializzatore)

È il costruttore. Prende i dati (Davide, Cariola) e li salva dentro l'oggetto. Senza di questo, l'oggetto non avrebbe "memoria" dei suoi dati.

2. __str__ (La Rappresentazione)

Senza questo metodo, se scrivessi print(davide), vedresti qualcosa di illeggibile come <__main__.Person object at 0x... >. Grazie a __str__, Python sa che deve stampare il nome e il cognome.

3. __eq__ (L'Uguaglianza)

Di base, Python considera due oggetti diversi anche se hanno gli stessi dati (perché occupano posti diversi in memoria). Con __eq__ gli diciamo: "Se nome e cognome coincidono, allora considera queste due persone come la stessa entità".

4. __setitem__ (Accesso a Dizionario)

Questo è un metodo avanzato. Ti permette di trattare l'oggetto come se fosse un Dizionario.
Invece di scrivere davide.name = "Luca", potresti scrivere davide["name"] = "Luca". È utilissimo per rendere le classi dinamiche e validate (infatti nel tuo codice lanci un'eccezione se la chiave è sbagliata).

5. __len__ (La Misura)

Definisce cosa significa la "lunghezza" di una persona. Nel tuo caso, hai deciso che è la somma delle lettere di nome e cognome. Quando scriverai len(davide), Python chiamerà automaticamente questo metodo.

6. __add__ (La Somma)

Intercetta l'operatore matematico +.

Quando Python vede persona1 + persona2, cerca questo metodo.

Nota importante: Il metodo riceve self (l'oggetto a sinistra del +) e other (l'oggetto a destra del +).
