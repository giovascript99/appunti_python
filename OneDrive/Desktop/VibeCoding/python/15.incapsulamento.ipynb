🛡️ Incapsulamento: Attributi Privati e Name Mangling

L'incapsulamento è la pratica di nascondere i dati interni di un oggetto per proteggerli da modifiche accidentali esterne.

1. Attributi Protetti (_nome)

In Python, un singolo trattino basso è solo una convenzione. Dice agli altri programmatori: "Ehi, questo attributo è interno, non toccarlo direttamente", ma tecnicamente è ancora accessibile.

2. Attributi Privati (__nome)

Quando usi due trattini bassi, Python attiva il Name Mangling. L'attributo viene "nascosto" cambiando il suo nome interno.

```python
class User:
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.__password = password
        
    def __str__(self):
        return f"{self.username} {self.email} - password: {self.password}"
    
davide = User('DavideC', 'davide.cariola@aulab.it', 12345)

davide.username = 'DaviCar'
davide.password = "qwerty"
print(davide.username)
print(davide.password)
print(davide)
print(davide._User__password)
```

3. Cos'è il Name Mangling?

Python non rende l'attributo davvero impossibile da raggiungere, ma lo rinomina internamente seguendo lo schema _NomeClasse__nomeAttributo.

Se provi ad accedere a davide.password, Python dice che non esiste.

Se scrivi davide._User__password, potresti ancora leggerlo (ma è caldamente sconsigliato!).

4. Perché usarlo?

Serve a evitare che le sottoclassi sovrascrivano accidentalmente attributi importanti o che l'utente finale modifichi dati sensibili (come un saldo bancario) senza passare per i metodi ufficiali (es. un metodo deposita()).

Conclusione: Usando davide.password = "qwerty", non hai violato la sicurezza della classe, hai solo aggiunto una "etichetta" esterna che si chiama nello stesso modo, ma la variabile originale __password è rimasta protetta all'interno.
