📂 Gestione dei File in Python

La gestione dei file permette a un programma di leggere dati da documenti esterni o di salvare informazioni in modo permanente sul disco.

```
Lettera       Descrizione                  Comportamento

 r             Read (Sola Lettura)          Default. Apre il file per leggere. Errore se il file non esiste.

 w             Write (Sola Scrittura)       Crea il file se non esiste. Attenzione: Se esiste, cancella tutto il contenuto precedente!

 x             Create (Esclusiva)           Crea il file e scrive. Errore se il file esiste già (evita sovrascritture accidentali).

 a             Append (In coda)             Apre il file per scrivere, ma aggiunge il testo alla fine senza cancellare nulla.

 +             Update (Aggiornamento)       Usato con altri (es. r+), permette sia di leggere che di scrivere nello stesso file.

 b             Binary (Binario)             Per file non testuali (immagini, video, eseguibili). Esempio: rb o wb.

 t             Text (Testuale)              Default. Apre il file interpretando i byte come caratteri (testo).

1. Apertura Manuale (open e close)

In questo metodo, apriamo il file e dobbiamo ricordarci esplicitamente di chiuderlo. Se il programma dovesse crashare prima del close(), il file potrebbe rimanere "appeso" o danneggiarsi.

````python
file = open("C:/Users/giope/OneDrive/Desktop/VibeCoding/python/pippo.txt", mode='r')

print(f"File {file.name} aperto in modalità {file.mode}")
print(f"Encoding: {file.encoding}")

file.close()
````

2. Il Contest Manager (with open)

Questa è la modalità raccomandata. Utilizzando la parola chiave with, Python gestisce automaticamente l'apertura e, soprattutto, la chiusura del file, anche se si verificano errori all'interno del blocco di codice.

````python
with open("C:/Users/giope/OneDrive/Desktop/VibeCoding/python/pippo.txt", mode='r') as file:
    print(f"File {file.name} aperto in modalità {file.mode}")
    print(f"Encoding: {file.encoding}")
    
print("Ho chiuso il file? ", file.closed)
````
