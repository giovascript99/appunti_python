📍 Il Puntatore: tell( ) e seek( )

Immagina un cursore invisibile che si muove mentre leggi.

file.tell( ): Ti dice esattamente in quale posizione (byte) si trova il cursore in questo momento.

file.seek(offset, whence): Sposta il cursore manualmente.

offset: Numero di posizioni di cui spostarsi.

whence: Punto di riferimento:

0: Inizio del file (Default).

1: Posizione attuale.

2: Fine del file.
