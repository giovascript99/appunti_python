✍️ Scrittura di File (mode='w', 'a', 'x')

Per scrivere su un file, dobbiamo cambiare la modalità (mode). Il metodo principale è file.write().

mode='w' (Write): Crea il file o sovrascrive quello esistente.

mode='a' (Append): Aggiunge testo alla fine del file senza cancellare nulla.

```python
with open('C:/Users/giope/OneDrive/Desktop/VibeCoding/python/to_write.txt', mode='w') as file:
    file.write('Hello World!')
```

🖼️ Manipolazione Binaria e Firme (Magic Numbers)

I file non testuali (come immagini JPG) devono essere aperti in modalità binaria ('rb' o 'wb'). Ogni tipo di file ha una "firma" (header e tail) che permette ai programmi di riconoscerlo.

```python
with open('C:/Users/giope/OneDrive/Desktop/VibeCoding/python/drone.jpg', mode='rb') as file:
    print(f"jpg signature head {file.read(4)}")
    file.seek(-2, 2)
    print(f"jpg signature tail: {file.read(2)}")
```

🕵️ Steganografia: Nascondere dati nei file

La steganografia consiste nel nascondere informazioni dentro altri file. Sfruttando il fatto che molti visualizzatori di immagini smettono di leggere dopo la firma di chiusura (tail), possiamo inserire dati personalizzati prima della fine del file.

```python
# stenografia
with open('C:/Users/giope/OneDrive/Desktop/VibeCoding/python/drone.jpg', mode='rb') as reader:
    with open('C:/Users/giope/OneDrive/Desktop/VibeCoding/python/drone_copy.jpg', mode='wb') as writer:
        writer.write(reader.read())
        reader.seek(-2, 2)
        tail = reader.read(2)
        writer.seek(-2, 2)
        writer.write(b"Info inserite")
        writer.write(tail)
```
