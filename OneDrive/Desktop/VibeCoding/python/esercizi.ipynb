```python
def somma_fino_a(n):
    if n == 1:
        return 1
    else:
        return n + somma_fino_a(n - 1)
    
somma_fino_a(4)
```

```python
numeri = list(range(1, 1001))
key = 725

def binary_search(numeri, key):
    print(numeri)
    if len(numeri) == 0:
        print("Numero non trovato")
        return False
    else:
        mid = len(numeri) // 2
        if numeri[mid] == key:
            print('Trovato')
            return True
        elif key < numeri[mid]:
            return binary_search(numeri[:mid], key)
        else:
            return binary_search(numeri[mid+1:], key)
        
binary_search(numeri, key)
        
    
```

```python

def filtra_maggiori(lista, soglia):
    risultato = []
    
    for numero in lista:
        if numero > soglia:
            risultato.append(numero)
            
    return risultato
    
numeri_test = [10, 5, 20, 8, 15, 20]
limite = 10

lista_filtrata = filtra_maggiori(numeri_test, limite)

print(f"Lista originale: {numeri_test}")
print(f"Numeri maggiori di {limite}: {lista_filtrata}")
```

```python
def counter_vocali(frase):
    vocali = "aeiou"
    counter = 0
    frase_min = frase.lower()
    for lettera in frase_min:
        if lettera in vocali:
            counter += 1
            
    return counter

frase = "Ciao mi chiamo Giovanni!"
risultato = counter_vocali(frase)
  
print(f"Nella frase '{frase}' ci sono {risultato} vocali.")          
```

```python
def inverti_stringa(stringa):
    if len(stringa) <= 1:
        return stringa
    
    return stringa[-1] + inverti_stringa(stringa[:-1])
    
frase = "Python"
risultato = inverti_stringa(frase)

print(f"La frase '{frase}' al contrario si legge '{risultato}'")
```

```python
def is_palindromo(testo):
    testo_pulito = testo.lower().replace(" ", "")
    testo_contrario = inverti_stringa(testo_pulito)
    return testo_pulito == testo_contrario
    
# --- TEST ---
parole = ["Anna", "Radar", "Python", "i topi non avevano nipoti"]

for p in parole:
    risultato = is_palindromo(p)
    print(f"'{p}' è palindromo? {risultato}")
```

```python
def conta_elementi(lista):
    conteggio = {}
    
    for elemento in lista:
        if elemento in conteggio:
            conteggio[elemento] += 1
        else:
            conteggio[elemento] = 1
    
    return conteggio

# --- TEST ---
frutta = ['mela', 'pera', 'mela', 'arancia', 'pera', 'mela']
risultato = conta_elementi(frutta)

print(f"Lista originale: {frutta}")
print(f"Conteggio finale: {risultato}")
```
